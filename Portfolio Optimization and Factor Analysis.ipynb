#**Portfolio Optimization**

```python
from numpy.random import random
```

```python
num_ports = 10000

all_weights = np.zeros((num_ports,len(tickers)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)


np.random.seed(123)
for ind in range(num_ports):
    weights = np.array(random(len(tickers)))
    weights = weights / np.sum(weights)
    all_weights[ind,:] = weights
    ret_arr[ind] = np.sum((ret.mean() * weights) *252)
    vol_arr[ind] = np.sqrt(np.dot(weights.T, np.dot(ret.cov() * 252, weights)))
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
sharpe_arr.max()
```

```python
index_maxsharpe= sharpe_arr.argmax()
index_maxsharpe
```

```python
all_weights[index_maxsharpe,:]
optimal_weights = all_weights[index_maxsharpe, :]
print("Optimal Weights for Each Stock:")
for ticker, weight in zip(tickers, optimal_weights):
    print(f"{ticker}: {weight:.2%}")
```

```python
max_sr_ret = ret_arr[index_maxsharpe]
max_sr_vol = vol_arr[index_maxsharpe]
max_sr_ret
```

#**Fama-French Regression**

```python
import pandas_datareader.data as web
```

```python
ff_factor = 'F-F_Research_Data_5_Factors_2x3_daily'
df_5factor_data = web.DataReader(ff_factor, 'famafrench', start=start_dt_port, end=end_dt_port)[0]
df_5factor_data.rename(columns={'Mkt-RF': 'MKT'}, inplace=True)
df_5factor_data=df_5factor_data/100
df_5factor_data
```

```python
portret_factor = pd.merge(portret,df_5factor_data,left_index=True,right_index=True)
portret_factor['XsRet'] = portret_factor['preturn'] - portret_factor['RF']
portret_factor
```

```python
from statsmodels.formula.api import ols
```

```python
model = ols("XsRet ~ MKT + SMB + HML + RMW + CMA", data = portret_factor)
results = model.fit()
print(results.summary())
```

Interpretation:

*   MKT: The portfolio is less volatile and moves with the market as the coefficient is close to 1
*   SMB: The coefficient is positive which indicates that the portfolio is behaving like  small cap stocks
*   HML: The positive coefficient indicates that the portfolio is focused on value as opposed to growth stock.
*   RMW:The positive coefficient indicates that the portfolio is focused on value as opposed to growth stock.
*   CMA: The negative coefficient indicates that my portfolio is on the aggressive side as opposed to being conservative
*   Analysis: The portfolio does not significantly outperform the Fama-French model. The p value of the intercept is larger than 0.05 which indicates that we are unable to conclude that the portfolio outperforms the model.
